1. What is multithreading in Python? Why is it used? Name the module used to handle threads in Python.

Ans:
    

Multithreading in Python refers to the concurrent execution of multiple threads (smaller units of a process) within a single Python program. Each thread runs independently and can perform its tasks concurrently with other threads. Multithreading is used to improve program performance, responsiveness, and to utilize multiple CPU cores effectively.

The module used to handle threads in Python is the threading module. It provides a higher-level interface for creating and managing threads in Python.

activeCount(): Returns the number of Thread objects currently alive and managed by the threading module.

currentThread(): Returns the Thread object representing the current thread.

enumerate(): Returns a list of all Thread objects currently alive and managed by the threading module.

Q2: Why is the threading module used? Write the use of the following functions (activeCount, currentThread, enumerate).

Ans:
    The threading module in Python is used to work with threads, allowing you to create and manage threads within your Python programs. Threads are lightweight sub-processes that can run concurrently, making it possible to perform multiple tasks simultaneously. The threading module provides a high-level interface for creating and controlling threads, making it easier to implement multithreading in your applications.

Here are the uses of the following functions in the threading module:

activeCount(): This function is used to determine the number of Thread objects that are currently alive and managed by the threading module. It returns an integer representing the count of active threads.

Example:import threading

# Create two threads
thread1 = threading.Thread(target=some_function)
thread2 = threading.Thread(target=another_function)

# Start the threads
thread1.start()
thread2.start()

# Get the count of active threads
count = threading.activeCount()
print(f"Number of active threads: {count}")

currentThread(): This function returns a Thread object representing the currently executing thread. You can use this function to obtain information about the current thread, such as its name or ID.

Example:import threading

def my_function():
    current_thread = threading.currentThread()
    print(f"Current thread name: {current_thread.name}")

thread = threading.Thread(target=my_function)
thread.start()

2. Explain the following functions (run, start, join, isAlive).
Ans:
    

run(): The run() method is the entry point for the thread's activity. It is invoked when you call the start() method on a Thread object. You can override this method in a custom thread class to define the thread's behavior.

start(): The start() method is used to start a thread's execution. It creates a new thread and calls the run() method in that thread. It should be called once per thread.

join(): The join() method is used to block the calling thread until the thread on which it's called (usually another thread created with start()) has finished its execution. It's used for synchronization between threads.

isAlive(): The isAlive() method checks whether the thread is currently executing (alive) or has finished its execution (dead).

3. Write a Python program to create two threads. Thread one must print the list of squares, and thread two must print the list of cubes.

Ans:
    
    import threading

def print_squares():
    for i in range(1, 6):
        print(f"Square of {i}: {i**2}")

def print_cubes():
    for i in range(1, 6):
        print(f"Cube of {i}: {i**3}")

# Create two thread objects
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")


4. State advantages and disadvantages of multithreading.
Ans:
    

##Advantages of Multithreading:

Improved Performance: Multithreading can utilize multiple CPU cores, making it suitable for CPU-bound tasks, improving execution speed.

Responsiveness: Multithreading enhances the responsiveness of applications, as threads can perform tasks concurrently without blocking the main thread.

Resource Sharing: Threads within the same process can share data and resources, making it efficient for tasks that require communication.

##Disadvantages of Multithreading:

Complexity: Writing multithreaded code can be complex and error-prone due to issues like race conditions and deadlocks.

Debugging: Debugging multithreaded programs can be challenging, as issues may not always be easily reproducible.

Overhead: Creating and managing threads incurs overhead, which can be significant if you create too many threads for short-lived tasks.

Resource Contention: Threads may compete for resources, potentially causing contention and performance degradation

5. Explain deadlocks and race conditions.

Ans:
    

Deadlock: A deadlock occurs when two or more threads are unable to proceed because each is waiting for the other(s) to release a resource or take some action. Deadlocks can lead to a program's complete halt, where no progress is possible. They typically occur in situations involving multiple locks or resources.

Example: Thread A holds Resource X and waits for Resource Y, while Thread B holds Resource Y and waits for Resource X. Both threads are stuck, and the program is deadlocked.

Race Condition: A race condition occurs when two or more threads access shared data concurrently, and the final outcome depends on the timing or order of execution. It leads to unpredictable behavior and can result in data corruption or unexpected results. Race conditions are often the result of improper synchronization.

Example: Two threads concurrently increment a shared counter. Without proper synchronization, the counter may end up with an incorrect value because the threads read and modify it simultaneously.

To avoid these issues, proper synchronization mechanisms, such as locks or semaphores, should be used in multithreaded programs to ensure that threads can access shared resources safely and prevent deadlocks and race conditions.
